```python
# 連接至資料庫 [B]
def connect_to_database(database_name):
    # 新增資料 [B]
    try: 
        conn  =  pymysql.connect ( host = '' ,  user = '' ,  passwd = "" ,  db = database_name ) 
        cur  =  conn.cursor()
        return conn, cur
    except Exception as e:
        print(e)
# 連接至資料庫 [B]
```

```python
# 新增資料 [B]
def insert_to_database(conn, cur, table_name, data_column, data_value):
    try: 
        conn  =  pymysql.connect ( host = ' ' ,  user = ' ' ,  passwd = " " ,  db = database_name ) 
        cur  =  conn.cursor()
        insert_sql = '''INSERT INTO `{}` 
                        ({}) 
                        VALUES ({})
                     '''.format(table_name, data_column, data_value)
        cur.execute(insert_sql)
        conn.commit()
        cur.close () 
        conn.close()
    except Exception as e:
        print(e)
# 新增資料 [E]
```

```python
# 從資料庫上撈取資料 [B]
def select_to_database(conn, cur, table_name, data_value):
    try: 
        select_sql = ''' SELECT `id`, `prediction_1`, `prediction_2`, `prediction_3`, `prediction_4`, `created_at`
                        FROM `{}` WHERE `created_at` = '{}'
                     '''.format(table_name, data_value)
        result_object = cur.execute(select_sql)
        results_values_list = cur.fetchall()
        result_key_list = [i[0] for i in cur.description]

        reframed = pd.DataFrame(results_values_list)
        reframed.columns = result_key_list
        return reframed
    except Exception as e:
        print(e)
# 從資料庫上撈取資料 [E]
```

```python
# 更新資料庫資料 [B]
def update_weather_data(database_name, update_sql):
    try: 
        conn  =  pymysql.connect ( host = ' ' ,  user = ' ' ,  passwd = " " ,  db = database_name ) 
        cur  =  conn.cursor() 
        cur.execute(update_sql)
        conn.commit()

        cur.close () 
        conn.close()
    except Exception as e:
        print(e)
# 更新資料庫資料 [E]
```
